# Neural Networks

**What are neural networks?**
Neural networks are algorithms in the AI paradigm that mimic the workings of an animal's brain. 

The human brain typically has [100 billion neurons](https://https://www.pnas.org/doi/10.1073/pnas.1201895109#:~:text=Remarkably%2C%20at%20an%20average%20of,as%20other%20primates%20(26).) that help the species thrive. Scientists observed this behaviour over decades and tried to model the way humans learn  throughout their lives. 

While neurons in the brain 'talk' to each other using chemical signals and electrical impulses, an artificial neuron is a mathematical function that can be 'taught' or 'trained' to classify data. Each network can have an input layer, multiple hidden layers, and an output layer.

Each node in a neural network can be represented by this formula:

**Types of Neural Networks**



*   Feedforward neural network: In feedforward neural networks, the flow of data is unidirectional, i.e, from the input nodes to the oputput nodes, passing through one or more hidden layers, if present. This kind of neural network is used for simple classification problems.

*   Perceptron: The perceptron is often referred to as the simplest neural network and classifies binary data. A more complex idea of the perceptron is extended to multilayer perceptron networks. It is referred to as the fully connected version of the feedforward neural network. Multilayer perceptrons are used for spam filtering and stock market analysis.


*   Radial basis functions: These neural networks have three layers, input, hidden, and output. The output layer linearly combines the radial basis functions of the parameters of the neurons and the inputs. Such neural networks are used for predicting time series.

*   Recurrent neural network: Recurrent neural networks loop data through the network for better results. The data is passed forward through the layers of neurons and is then sent backwards in a loop. The nodes of such a network retain information in their memory and fine-tune the prediction. Recurrent neural networks are used for sequential data and language translations.



*   Modular neural network: A modular neural network combines multiple neural networks to perform a complex task. Each neural network is independent and tackles a specific subtask of the main task. A moderating layer then collects these outputs and produces the final result.

# Beginner

In this section, we will design a simple neural network.



Let's consider the equation for the straight line: y=mx+c

For a function f(x), let m = 5, c = 3

Hence, f(x): y=5x+3

We define the dataset as follows. We will feed this to our model.

```python
x = np.array([-4, -2, 0, 3, 4, 6, 8, 9], dtype = float)
y = np.array([-17, -7, 3, 18, 23, 33, 43, 48], dtype = float)
```

We define our model, beginner. It is a Sequential model, taken from Keras.

units = 1 represents the usage of one neuron and input_shape represents the number of inputs.

Hence, beginner is a Sequential model with one layer containing one neuron that takes one input.



Sequential models are machine learning models that are built like a linear stack of layers. 



```python
beginner = models.Sequential([layers.Dense(units = 1, input_shape = [1])])
```

Optimizers are mathematical functions modelled into algorithms, that minimize errors and maximize efficiency. We are using SGD or Stochastic Gradient Descent as the optimizer for this model.

The SGD updates the parameters of a model one after the other. Although it can be expensive computationally and generates high variance, SGD is efficient as it arrives at the result faster due to frequent updates.

A loss function is the evaluation criteria for a machine learning model. It represents how good or bad our model is.

There are various loss functions. We will use two in this tutorial.

Mean squared error (MSE): MSE is a basic loss function that squares the difference between the real value and the predictions of a model. 

Sparse categorical crossentropy: Sparse Categorical Crossentropy is a used for models with multi-class functionalities. This loss function works for output labels with integer values.

beginner.compile compiles beginner.

```python
beginner.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

This step is to train the model. Epochs are the number  of cycles the model looks through the data. Hence, beginner looks through the data 5 times here. You can decide the number of epochs based on your requirements.

```python
beginner.fit(x, y, epochs = 5)
```

Our model has now been trained!

We now predict the value of y for x=13.0

For this function, 
y=5*13+3 = 68

```python
predict_value = beginner.predict([13.0])
```

```python
print(predict_value)
```

The predicted value is very close to the real value. Our model is thus a good one!

# Loading the MNIST Handwriting Dataset

We will use the MNIST handwriting dataset with 70k images and labels of handwritten digits (0-9). This dataset is further divided into a training set of 60k images and a testing set of 10k images. Each image in this dataset is 28x28 pixels.

Much research has been done on this MNIST dataset and computer scientists, after years of work, could finally achieve 0.18% error in 2018. This means human-like accuracy, in some cases, may be even better than that. The model that achieved this was a system of fully connected recurrent convolutional neural network having the properties of all the three types.

We reference the dataset with a variable, data using the keras.dataset.mnist call that references the dataset.

```python
data = keras.datasets.mnist
```

We now load the dataset into two sets of variables- features and labels, and test_features and test_variables. The first set containing features or images and their labels as 'labels' consist of 60k 28x28 images. This is the training set that the model will have access to and will evaluate during the training process.

The testing set comprises of 10k images and labels that we are calling test_features and test_labels. The model will not have access to this data before it is fully trained. Once the model has been trained, we will provide the testing dataset as input to ensure our model is not suffering from overfitting. 

Overfitting is a machine learning 'disease' that models can suffer from if they are perfect against their training dataset. This would mean, while the model is great at predicting data from the dataset it has seen already, it cannot recognise data that it has never seen before and performs poorly against its testing dataset.

Similarly, underfitting occurs when the model is unable to understand the pattern of the training data and performs poorly.

```python
(features, labels), (test_features, test_labels) = data.load_data()
```

The shape function shows the shape of the data.

```python
features.shape
```

```python
test_features.shape
```

Next, we want to analyse the dataset and look at the images and their labels. To do this, we use matplotlib and the code snippet given below.

We see four images from the training dataset, along with their labels.

```python
for i in range(4):
  plt.subplot(320+1+i)
  plt.title((labels[i]))
  plt.imshow(features[i])
  plt.subplots_adjust(bottom=0.08, top=1.8)
  plt.axis("off")
```

Why are the output images coloured? This is because matplotlib uses the colour palette, viridis as default. To visualise them as gray-scale images, we use cmap as follows.

```python
for i in range(4):
  plt.subplot(320+1+i)
  plt.title((labels[i]))
  plt.imshow(features[i], cmap=plt.get_cmap('gray'))
  plt.subplots_adjust(bottom=0.08, top=1.8)
  plt.axis("off")
```

# Changing image values from 0-255 to 0-1

```python
features[0]
```



If we look at the numeric data that makes up the images, we will see it consists of 0-255. To ensure our model performs better and faster, we will transform these images to consist of numeric data between 0 and 1. The easiest way to do that is to divide both features and test_features by 255.

```python
features = features/255.0
test_features = test_features/255.0
```

```python
features[0]
```

We are now ready to create our model!

We name it 'advanced' and define it as a Sequential model with two layers.

The Dense layers of the Convolutional Neural Network or CNN can only accept one-dimensional inputs. Hence, we use the Flatten function to transform 28x28 images into 1D data.


Next, our model's layers are defined by two Dense layers. Here, we must talk about activation functions.

An activation function is simply the function that processes an input to produce an output. Hence, the inputs received by the first layer is processed using ReLU, producing an output that then becomes the input to the second layer that is processed by Softmax.

ReLU: Rectified Linear Unit or ReLU is an activation function that produces 0 for negative inputs and the number itself for positive inputs. ReLU is popular for its ease of use and the higher accuracy it helps the model obtain.

Softmax: Softmax is used for multi-class machine learning problems. It assigns probabilities to all classes present in the dataset and selects the class with the highest probability. All the probabilities given by softmax add upto 1.0. Softmax is used with the output layer of a neural network and must essentially have as many neurons as the number of classes. In our case, it has 10 neurons for the 10 classes (0-9).

```python
advanced = models.Sequential([layers.Flatten(input_shape = (28, 28)),
                         layers.Dense(128, activation = 'relu'),
                         layers.Dense(10, activation = tf.nn.softmax)])
```

# We compile our advanced model

```python
advanced.compile(optimizer = 'adam',
                 loss = 'sparse_categorical_crossentropy',
                 metrics = ['accuracy'])
```

# We train our model

```python
advanced.fit(features, labels, epochs = 5)
```

```python
loss_result, accuracy_result = advanced.evaluate(test_features, test_labels)
```

We want to check the loss and accuracy value of our model.

```python
print("Resultant Loss: ", loss_result)
print("Resultant Accuracy: ", accuracy_result)
```

The loss is 0.08 and the accuracy is 97.5%
